```python
%pylab inline
from scipy.special import erf
import numpy as np

k=(10**-6) #Diffusivity
ts_tm=273-873 #Surface temperature - Mantle Temperature
#pasar_a_sec=(1/(100*365*24*3600))
#v1=0.5*pasar_a_sec
v1= 0.000000000158549

```

```python
nx, ny = 500, 500
x = np.linspace(-500, 500, nx)
y = np.linspace(-500,0, ny)
#xv, yv = meshgrid(x, y, indexing='ij')
xv, yv = meshgrid(x, y)
z= erf(yv/(2*(sqrt((k*abs(xv)/v1)))))
Temp=(z*(ts_tm))+273
```

```python
CS=contour(xv,yv,Temp)
CB = colorbar(CS)
```

```python
def TBL(v,x):
    tbl=2.32*(sqrt((k*x)/v))
    return tbl  

thermal_boundary = []
x2 = np.linspace(0, 500, nx)
for i in range(500):
    tbl_l=-TBL(v1,i)
    thermal_boundary.append(tbl_l)
#print result
plot(x2,thermal_boundary)
    
    
    
        
```

```python
#Ocean floor Bathymetry

def w(v,x):
    mantle_density=3300
    water_density=1000
    delta_T=1300
    alpha=1.6*(10**-5)
    W=((2*mantle_density*alpha*(delta_T))/(mantle_density-water_density))*sqrt((k*x)/(3.14*v))
    return W

bathimetry = []

for i in range(500):
    bath=w(v1,i)
    bathimetry.append(bath)
```

```python
plt.plot(x2, thermal_boundary,'r')
plt.plot(x2, bathimetry,'b')
plt.show()
```

## Temperature as a function of x and y for three different velocities

```python
%pylab inline
from scipy.special import erf

from scipy.special import erfinv

import numpy as np
        
k=(10**-6) #Diffusivity

ts=273 #Temp in Kelvin
tm=873 #TEmp in Kelvin
ts_tm=273-873 #Surface temperature - Mantle Temperature

#velocities
v1= 0.000000000158549 #Slow velocity m/yr
v2= 0.0000000012683917 #Moderate velocity m/yr
v3= 0.0000000031709792 #Fast velocity m/yr

```

```python
#HW 1.a
#X and y grid for implementing the function
nx, ny = 500, 500
x = np.linspace(-500, 500, nx)
y = np.linspace(-500,0, ny)
xv, yv = meshgrid(x, y)

#error function for each velocity
zv1= erf(yv/(2*(sqrt((k*abs(xv)/v1)))))
Temp_v1=(zv1*(ts_tm))+273
zv2= erf(yv/(2*(sqrt((k*abs(xv)/v2)))))
Temp_v2=(zv2*(ts_tm))+273
zv3= erf(yv/(2*(sqrt((k*abs(xv)/v3)))))
Temp_v3=(zv3*(ts_tm))+273
```

```python
CS=contour(xv,yv,Temp_v1)
plt.title('T(x,y) slow v')
plt.xlabel('x')
plt.ylabel('y')
CB = colorbar(CS)
```

```python
CS=contour(xv,yv,Temp_v2)
plt.title('T(x,y) medium v')
plt.xlabel('x')
plt.ylabel('y')
CB = colorbar(CS)
```

```python
CS=contour(xv,yv,Temp_v3)
plt.title('T(x,y) fast v')
plt.xlabel('x')
plt.ylabel('y')
CB = colorbar(CS)
```

## Each T(x,y) varies in a different way according to the velocity of the ridge. As we notice the relationship s 1/srt (v), so  For less velocity, we have higher changes on T, and for faster velocities, we have slowest, or more distanced canges in T.

# 1.b. Thermal Boundary layers as function of x for 3 velocities

```python
#HW 1.b
def TBL(v,x):
    tbl=2.32*(sqrt((k*x)/v))
    return tbl  
#Thermal boundary for v1
thermal_boundaryv1 = []
x2 = np.linspace(0, 500, nx)
for i in range(500):
    tbl_l=-TBL(v1,i)
    thermal_boundaryv1.append(tbl_l)

plot(x2,thermal_boundaryv1)
    
plt.title('TBL slow v')
plt.xlabel('x')
plt.ylabel('TBL')
    
    
        
```

```python
thermal_boundaryv2 = []
#Thermal boundary for v2
x2 = np.linspace(0, 500, nx)
for i in range(500):
    tbl_l=-TBL(v2,i)
    thermal_boundaryv2.append(tbl_l)

plot(x2,thermal_boundaryv2)

plt.title('TBL medium v')
plt.xlabel('x')
plt.ylabel('TBL')
    
```

```python
thermal_boundaryv3 = []
#Thermal boundary for v2
x2 = np.linspace(0, 500, nx)
for i in range(500):
    tbl_l=-TBL(v3,i)
    thermal_boundaryv3.append(tbl_l)

plot(x2,thermal_boundaryv3)
plt.title('TBL fast v')
plt.xlabel('x')
plt.ylabel('TBL')
    
    
```

# 1.c Ocean floor Bathymetry- Plate thickness

```python
#HW 1.c
#Ocean floor Bathymetry

def w(v,x):
    mantle_density=3300
    water_density=1000
    delta_T=1300
    alpha=1.6*(10**-5)
    W=((2*mantle_density*alpha*(delta_T))/(mantle_density-water_density))*sqrt((k*x)/(3.14*v))
    return W

bathimetryv1 = []

for i in range(500):
    bath1=w(v1,i)
    bathimetryv1.append(bath1)
    
bathimetryv2 = []

for i in range(500):
    bath2=w(v2,i)
    bathimetryv2.append(bath2)
    
bathimetryv3 = []

for i in range(500):
    bath3=w(v3,i)
    bathimetryv3.append(bath3)
```

```python
plt.plot(x2, thermal_boundaryv1,'r')
plt.plot(x2, bathimetryv1,'b')
plt.title('Bathimetry slow v')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
plt.plot(x2, thermal_boundaryv2,'r')
plt.plot(x2, bathimetryv2,'b')
plt.title('Bathimetry medium v')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
plt.plot(x2, thermal_boundaryv3,'r')
plt.plot(x2, bathimetryv3,'b')
plt.title('Bathimetry fast v')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

# 1.d. Relaxation time as a function of T

```python
#Punto 1d
sigma=1*(10**6) #pa crustal stress at 3km
A=8*(10**3) #Pa
n=3.1
Q=240
R=8.3*(10**-3)
E=75*(10**6)#MPa oc crust E is 100GPa

temp_occrust = np.linspace(1473, 273, 500)

#print temp_occrust
def Mu(T):
    mu=(((sigma**(1-n))*(exp(Q/(R*(T-273)))))/(2*A))
    return mu


rel_time= []

for i in range(500):
    
    Mhu=Mu(temp_occrust[i])
    rel_time.append(Mhu)
    
from matplotlib import pyplot
plot(temp_occrust,rel_time)
pyplot.yscale('log')
```

# Elastic viscuous transition temperature for 3 velocities

## Between 300 and 500 There is a transition. We have a change of behaviour. We have high relaxation times for elastic behaviours, (low Temp) We have low relaxation tmes for more viscous materials (high Temp).  

```python
n=500
T = np.linspace(300, 500, n)
Xn = np.linspace(300, 500, n)

y_thermal= []

for i in range(n):
    y=(erfinv((T[i]-300)/(200))*(2*(sqrt((k*abs(Xn[i])/v1)))))
    y_thermal.append(-y)
```

```python
plt.plot(Xn,y_thermal)
plt.plot(x2, thermal_boundaryv1,'r')
plt.plot(x2, bathimetryv1,'b')
plt.title('Elastic thickness Transition(y)  with lithospheric profile slow v')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
y_thermal2= []

for i in range(n):
    y=(erfinv((T[i]-300)/(200))*(2*(sqrt((k*abs(Xn[i])/v2)))))
    y_thermal2.append(-y)

plt.plot(Xn,y_thermal2)
plt.plot(x2, thermal_boundaryv1,'r')
plt.plot(x2, bathimetryv1,'b')
plt.title('lastic thickness Transition(y) with lithospheric profile medium v ')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
y_thermal3= []

for i in range(n):
    y=(erfinv((T[i]-300)/(200))*(2*(sqrt((k*abs(Xn[i])/v3)))))
    y_thermal3.append(-y)
    
plt.plot(Xn,y_thermal3)
plt.plot(x2, thermal_boundaryv1,'r')
plt.plot(x2, bathimetryv1,'b')
plt.title('lastic thickness Transition(y)  with lithospheric profile fast v ')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

# Asvelocities increase and mantle convection becomes more vigorous, we can see that isotherms generated from the ridges are more separated. ALso the Themal Boundary Layers would be at a shallower level, than with slow spreading ridges. Bathimetry isnt varying that much with the increase on velocity. Finally the elastic-viscuous transition would be given at shallower depths, as shown on the profiles.
